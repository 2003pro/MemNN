```python
import numpy as np
import scipy
from functools import partial
from scipy.optimize import minimize
from sklearn.feature_extraction.text import *
from sklearn.preprocessing import *

TRAIN_FILE='data/en/qa1_single-supporting-fact_test.txt'
TEST_FILE='data/en/qa1_single-supporting-fact_test.txt'
GAMMA = 0.1
```

```python
def phi_x(x, vectorizer):
    lenW = len(vectorizer.vocabulary_)
    m = vectorizer.transform([x['text']]).toarray().T
    if x['type'] == 'q':
        retval = np.concatenate([m, np.zeros((2*lenW, 1))], axis=0)
    else:
        retval = np.vstack((np.zeros((lenW, 1)), m, np.zeros((lenW, 1))))
    retval = np.matrix(retval)
    return retval

def phi_y(y, vectorizer):
    lenW = len(vectorizer.vocabulary_)
    retval = np.concatenate([np.zeros((2*lenW, 1)), vectorizer.transform([y['text']]).toarray().T])
    retval = np.matrix(retval)
    return retval

def s(x, y, U, vectorizer):
    retval = 0
    for xx in x:
        A = phi_x(xx, vectorizer).T
        B = phi_y(y, vectorizer)
        result = A * U.T * U * B
        retval += result[(0,0)]
    return retval

def loss(x, f1, f2, r, memory_list, U_O, U_R, vectorizer):
    if f1 is not None:
        f1 = memory_list[f1-1]
    if f2 is not None:
        f2 = memory_list[f2-1]
    sO = partial(s, U=U_O, vectorizer=vectorizer)
    sR = partial(s, U=U_R, vectorizer=vectorizer)
    best = None
    bestVal = None
    for f_bar in memory_list:
        val = sO([x], f_bar)
        if bestVal is None or val > bestVal:
            best = f_bar
            bestVal = val
    m_o1 = best

    best = None
    bestVal = None
    for f_bar in memory_list:
        val = sO([x, m_o1], f_bar)
        if bestVal is None or val > bestVal:
            best = f_bar
            bestVal = val
    m_o2 = best         

    retval = 0
    #indices = np.arange(len(memory_list))
    #np.random.shuffle(indices)
    #memory_list = np.array(memory_list)[indices[:2]]
    for f_bar in memory_list:
        if f_bar == f1:
            continue
        retval += max(0, GAMMA - sO([x], f1) + sO([x], f_bar))
    
    if f2 is not None:
        for f_bar in memory_list:
            if f_bar == f2:
                continue
            retval += max(0, GAMMA - sO([x, m_o1], f2) + sO([x, m_o1], f_bar))
            
    for r_bar in vectorizer.vocabulary_:
        if r_bar == r:
            continue
        retval += max(0, GAMMA - sR([x, m_o1, m_o2], {'text':r}) + sR([x, m_o1, m_o2], {'text':r_bar}))
    return retval
```

```python
def train(fname):    
    lines = []
    for line in open(fname):
        id = int(line[0:line.find(' ')])
        line = line.strip()
        line = line[line.find(' ')+1:]        
        if line.find('?') == -1:
            lines.append({'type':'s', 'text': line})
        else:
            idx = line.find('?')
            tmp = line[idx+1:].split('\t')
            lines.append({'id':id, 'type':'q', 'text': line[:idx], 'answer': tmp[1].strip(), 'refs': [int(x) for x in tmp[2:][0].split(' ')]})
    vectorizer = CountVectorizer()
    vectorizer.fit([x['text'] for x in lines])
    lenW = len(vectorizer.vocabulary_)
    
    def objective(x):
        tmp = x.reshape((200, 3*lenW))
        U_O = tmp[range(100),:]
        U_R = tmp[range(100,200),:]
        
        retval = 0
        for i,line in enumerate(lines):
            if line['type'] == 'q':
                refs = line['refs']
                f1 = refs[0]
                f2 = refs[1] if len(refs) > 1 else None
                r = line['answer']
                id = line['id']
                memory_list = lines[i-id+1:i]
                retval += loss(line, f1, f2, r, memory_list, U_O, U_R, vectorizer)
        
        return retval

    x0 = np.concatenate([np.random.rand(100, 3*lenW), np.random.rand(100, 3*lenW)], axis=0)
    x0 = x0.ravel()
    res = minimize(objective, x0, method='BFGS', options={'disp':True})
    print res
    
    
```

```python
train(TRAIN_FILE)
```
